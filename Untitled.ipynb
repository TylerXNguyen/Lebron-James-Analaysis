### WEBSCRAPPING CYBER SECURITY DATA

```python
url = "https://en.wikipedia.org/wiki/LeBron_James"
type_ = {"class": "wikitable sortable"}
page = requests.get(url)
soup = BeautifulSoup(page.text, "lxml")
table = soup.find("table", type_) # (change this) depend on table
# get headers
headers = []
for col in table.find_all('th'):
    # gets each col label (title) from table
    title = col.text
    headers.append(title.strip())
# create df from headers and data
df = pd.DataFrame(columns = headers)
for row in table.find_all('tr')[1:]:
    # getting data from each rows after the header
    try:
        data = row.find_all('td')
        row_data = [td.text.strip() for td in data]
        df.loc[len(df)] = row_data # appends new row to df
    except:
        0
df.head()
```

### Data Cleaning

```python
# Removes wiki symbols
for col in df.columns:
    df[col] = df.apply(lambda x: x[col].replace('*', ''), axis=1) # removes * from values
    df[col] = df.apply(lambda x: x[col].replace('†', ''), axis=1) # removes † from records col
    
# Fix year to first term
df["Year"] = df.apply(lambda x: x["Year"][:-3], axis=1) # removes sequential year
```

```python
def cleanNumerical(table, col, fix = -1):
    # PURPOSE:
        # Cleans an intended numerical column and shows info about how to replace bad values
        # Also removes commas from values
    # INPUT: 
        # table (DF) - table to take
        # col (string) - value for the column to use
        # fix (int/double) - default -1, if fix set other than -1, cleans bad values to integer with parameter
    # OUTPUT:
        # Prints Mean and median of good values
        # Prints bad values not integer

    try:
        table[col] = table.apply(lambda x: x[col].replace(',', ''), axis=1) # removes commas from records col
        print("Converted comman values")
    except:
        print("")
    success = []
    failures = []
    for i in range(len(table)):
        try:
            if ("." in table.loc[i, col]):
                # For decimal (float) values
                table.loc[i, col] = float(table.loc[i, col])
            else:
                # For int values
                table.loc[i, col] = int(table.loc[i, col])
            success.append(table.loc[i, col])
        except:
            if fix != -1:
                table.loc[i, col] = fix
            else:
                failures.append(table.loc[i, col])

    print("Mean of correct values:", statistics.mean(success))
    print("Median of correct values:", statistics.median(success))
    if len(failures) > 0:
        print("Items not numerical:")
        for i in set(failures):
            print(i)
    else:
        print("No more bad format values")
    print("")
```

```python
# checking valid values of numeric cols
numericCols = df
numericCols = numericCols.drop(columns = ["Team"]) # cols of only numeric

# Goes through each value
for cols in numericCols.columns:
    # each cols
    for rows in range(len(numericCols)):
        # each row
        if (numericCols.loc[rows, cols][0] == "."):
            # Add 0 to percentage when necessary
            numericCols.loc[rows, cols] = str(0) + numericCols.loc[rows, cols]
            
# puts into numeric format
for cols in numericCols.columns:
    cleanNumerical(numericCols, cols)
```

```python
# finalize data cleaning to df2
df2 = numericCols
df2["Team"] = df["Team"]
df2.head()
```

### Train a Regression Model

```python
# Separate features and labels
X = df2[["Year", "GP", "GS", "MPG", "FG%", "3P%", "FT%", "RPG", "APG", "SPG", "BPG"]].values # change this
yName = "PPG" # Change this
y = df2[yName].values
# print('Features:',X[:10], '\nLabels:', y[:10], sep='\n')
# Split data 70%-30% into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
# print ('Training Set: %d rows\nTest Set: %d rows' % (X_train.shape[0], X_test.shape[0]))
# Fit a linear regression model on the training set
model = LinearRegression().fit(X_train, y_train)
# print(model)
predictions = model.predict(X_test)
np.set_printoptions(suppress=True)
# print('Predicted labels: ', np.round(predictions)[:10])
# print('Actual labels   : ' ,y_test[:10])
# Print graph
%matplotlib inline
plt.scatter(y_test, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title(str(yName) + " Predictions") # Change this
# overlay the regression line
z = np.polyfit(y_test.astype("float"), predictions, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test), color='magenta')
plt.show()
# Residual metrics
mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = r2_score(y_test, predictions)
print("R2:", r2)
```

#### Commonly used residual evaluation metrics
- Mean Square Error (MSE): The mean of the squared differences between predicted and actual values. This yields a relative metric in which the smaller the value, the better the fit of the model
- Root Mean Square Error (RMSE): The square root of the MSE. This yields an absolute metric in the same unit as the label (in this case, numbers of rentals). The smaller the value, the better the model (in a simplistic sense, it represents the average number of rentals by which the predictions are wrong!)
- Coefficient of Determination (usually known as R-squared or R2): A relative metric in which the higher the value, the better the fit of the model. In essence, this metric represents how much of the variance between predicted and actual label values the model is able to explain.
